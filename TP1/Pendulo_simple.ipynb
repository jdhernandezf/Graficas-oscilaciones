```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive
```

```python
#t1=np.linspace(0,20,1000)
#print(t)
#theta=np.pi/4
g=9.8
m=1
lo=25
v0=15
#angulo=60
n=2500

#--------------------------------------------------------------------------------------
#-----------------------------SOLUCIÓN NUMERICA

#Ec difernecial
def theta(t,th):
    return -(g/lo)*np.sin(th)

#Método numerico
def RoKuy(t0,th0,vth0,h,n,ang):
    #print('y(',t,')=',y)
    t=np.zeros(n+1)
    vth=np.zeros(n+1)
    th=np.zeros(n+1)
    t[0]=t0
    vth[0]=vth0
    th[0]=th0
    for l in range(n):
        l1=h*theta(t[l],th[l])
        k1=h*vth[l]
        l2=h*theta(t[l]+h/2,th[l]+l1/2)
        k2=h*(vth[l]+l1/2)
        l3=h*theta(t[l]+h/2,th[l]+l2/2)
        k3=h*(vth[l]+l2/2)
        l4=h*theta(t[l]+h,th[l]+l3)
        k4=h*(vth[l]+l3)
        t[l+1]=t[l]+h
        tf=t[l+1]
        vth[l+1]=vth[l]+(1/6)*(l1+2*l2+2*l3+l4)
        #x[k+1]=x[k]+(h/6)*(k1+2*k2+2*k3+k4)
        th[l+1]=th[l]+(1/6)*(k1+2*k2+2*k3+k4)
        #print('y(',t[k+1],')=',y[k+1])
        
    #if th0>np.pi/12:
    #    plt.plot(t,th, label=r"numerica $\theta=%.1f$"%ang,color="green")
    #else:
    plt.plot(t,th, label=r"numerica $\theta=%.1f$"%ang,color="blue")
    
    plt.legend(loc=1)
    return th,vth

def Grafica(angulo):
    ang=angulo
    th0=angulo*np.pi/180
    th,vth=RoKuy(0,th0,0,0.008,n,ang)

    #th,vth=RoKuy(0,th0,0,0.00008,n,ang)

    #------------------------------------------------------------------------------------

    #Solución análitica
    delta=0
    def x(t):
        return th0*np.cos((g/lo)**0.5*t-delta)

    t=np.linspace(0,20,100)

    plt.title("Posición, velocidad y análitica")
    plt.xlabel("Tiempo (s)")
    plt.ylabel("Angulo ($rad$)")
    plt.plot(t,x(t),"--",label="análitica",color="green")
    plt.xlabel('tiempo ($s$)')
    plt.legend(loc=1)

    #------------------------------------------------------------------------------------

    plt.grid()
    plt.show
```

```python
interactive_plot = interactive(Grafica,angulo=(1,70,0.1))
interactive_plot
```
