```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Condiciones iniciales
#----------------------------------------------------------------------

Ax=1
Ay=1
fi=np.pi/2
wx=2
wy=3

#----------------------------------------------------------------------

#Calcular

def sin1(t):
    return Ax*np.sin(wx*t)

def sin2(t):
    return Ay*np.sin(wy*t+fi)

##################Pasos para graficar la variable independiente#############
T=10                   #Hasta donde graficar en el tiempo
n=1000                      #Numero de puntos a graficar
t=np.linspace(0,T,n+1)

##################construi la grafica####################
#plt.figure(figsize=figsize,dpi=dpi)
plt.plot(sin1(t),sin2(t))
plt.title("y vs x")
plt.ylabel("Oscilación en y")
plt.xlabel("Oscilación en x")
plt.show
```

```python
rows = 6
cols = 3
axes=[]
fig=plt.figure(figsize=(9,15))

for a in range(rows*cols):
    axes.append( fig.add_subplot(rows, cols, a+1) )
    #print(rows)
    if (a+1)<=3:
        wx=1
        wy=1
    elif (a+1)<=6:
        wy=2
    elif (a+1)<=9:
        wy=3
    elif (a+1)<=12:
        wx=2
    elif (a+1)<=15:
        wx=3
        wy=4
    elif (a+1)<=18:
        wy=5
    
    if ((a+3)%3)==0:
        fi=0
    elif(a+2)%3==0:
        fi=np.pi/4
    elif(a+1)%3==0:
        fi=np.pi/2
    plt.plot(sin1(t),sin2(t))
    subplot_title=(str(a+1))
    axes[-1].set_title(subplot_title)  
fig.tight_layout()    
plt.show()
```
